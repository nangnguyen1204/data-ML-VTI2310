# Mean bias error

image.png

```python
def tinh_MBE(actual, predicted):
  if len(actual) != len(predicted):
    raise ValueError("Chieu dai actual va predicted khong khop")

  mbe = np.mean(np.sum(actual - predicted))

  return mbe
```

```python
print(tinh_MBE(actual, predicted))
```

# Mean absolute error

image.png

```python
def tinh_MAE(actual, predicted):
  if len(actual) != len(predicted):
    raise ValueError("Chieu dai actual va predicted khong khop")

  mae = np.mean(np.sum(np.abs(actual - predicted)))

  return mae
```

```python
print(tinh_MAE(actual, predicted))
```

# Mean square error

image.png

```python
def tinh_MSE(actual, predicted):
  if len(actual) != len(predicted):
    raise ValueError("Chieu dai actual va predicted khong khop")

  square_error = (actual - predicted)**2

  mse = np.mean(square_error)

  return mse
```

```python
print(tinh_MSE(actual,predicted))
```

# Log cosh loss

image.png

image.png

```python
def tinh_logcosh(actual, predicted):
  if len(actual) != len(predicted):
    raise ValueError("Chieu dai actual va predicted khong khop")

  cosh = np.cosh(predicted - actual)

  logcosh = np.mean(np.log(cosh))

  return logcosh
```

```python
print(tinh_logcosh(actual, predicted))
```

#Binary Cross Entrophy (BCE)

image.png

```python
def tinh_BCE(actual, predicted):
  if len(actual) != len(predicted):
    raise ValueError("chieu dai cua actaul va predicted khong khop")

  a = actual*(np.log(predicted))
  b = (1 - predicted)*(np.log(1 - predicted))

  bce = -np.mean(a + b)

  return bce
```

```python
print(tinh_BCE(actual, predicted))
```
