##Nguyễn Công Năng

```python
import numpy as np
```

```python
x = np.random.randint(1,100,100)
print(x)
print(type(x))
```

```python
import matplotlib.pyplot as plt
plt.plot(x)
plt.title('Đồ thị gốc X')
plt.show()
```

```python
def find_max(ds):
  max_value = ds[0]
  for num in ds:
    if num > max_value:
      max_value = num
  return max_value
list_data = x
max_value_data = find_max(list_data)
print("Giá trị lớn nhất trong dãy số là : ", max_value_data)
```

```python
a = max(list_data)
a
```

```python
def find_min(ds):
  min_value = ds[0]
  for num in ds:
    if num < min_value:
      min_value = num
  return min_value
min_value_data = find_min(list_data)
print("Giá trị nhỏ nhất trong dãy số là : ", min_value_data)
```

```python
def normalization(list_data):
  normal_list = []
  for i in list_data:
    a = (i - min_value_data)/(max_value_data - min_value_data)
    normal_list.append(a)
  return normal_list
a = normalization(list_data)
a
```

```python
type(list_data)
```

```python
# Tính normalization

list_data_nor = (list_data - min_value_data)/(max_value_data - min_value_data)
list_data_nor
```

```python
#đồ thị sau khi normalization

plt.plot(list_data_nor)
plt.title('X sau khi Normalization')
plt.show()
```

```python
# tinhs mean
def tinh_mean_c1(list):
  sum = 0
  for i in list:
    sum += i
    mean = sum/len(list)
    return mean

tinh_mean_c1(x)
```

```python
def tinh_mean_c2(list):
  sum = np.sum(list)
  mean = sum/len(list)
  return mean

tinh_mean_c2(x)
```

```python
def tinh_mean_c3(list):
  size = np.size(list)
  sum = np.sum(list)
  mean = sum/size
  return mean

tinh_mean_c3(x)
```

```python
# Tinh do lech chuan standard
def standard(data):
  size = np.size(data)
  y = np.sum((data - tinh_mean_c3(data))**2)
  z = y/ size
  std = np.sqrt(z)

  return std

result = standard(x)
print(result)
```

```python
# tinh standardization

def Standardlization(data):
  stand_data = (data - tinh_mean_c3(data))/standard(data)
  return stand_data

result1 = Standardlization(x)
result1
```

```python
# Do thi sau khi standardlization
plt.plot(result1)
plt.title("X sau khi standardization")
plt.show()
```
