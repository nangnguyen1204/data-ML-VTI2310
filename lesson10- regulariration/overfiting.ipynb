```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# To d liu gi nh
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.randn(80) * 0.1

# Sp xp li d liu
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]

# To m hnh Linear Regression n gin
model_simple = LinearRegression()
model_simple.fit(X, y)
y_pred_simple = model_simple.predict(X_test)

# To m hnh Linear Regression qu phc tp (s dng a thc bc cao)
poly_features = PolynomialFeatures(degree=30)
X_poly = poly_features.fit_transform(X)
model_complex = LinearRegression()
model_complex.fit(X_poly, y)
X_test_poly = poly_features.transform(X_test)
y_pred_complex = model_complex.predict(X_test_poly)

# V biu  d liu v d on ca c hai m hnh
plt.scatter(X, y, s=20, label='D liu thc t')
plt.plot(X_test, y_pred_simple, label='M hnh Linear Regression n gin', color='g')
plt.plot(X_test, y_pred_complex, label='M hnh Linear Regression qu phc tp', color='r')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('M hnh Linear Regression b overfitting')
plt.show()

# o  tt ca m hnh bng Mean Squared Error
mse_simple = mean_squared_error(y, model_simple.predict(X))
mse_complex = mean_squared_error(y, model_complex.predict(X_poly))

print("MSE (n gin):", mse_simple)
print("MSE (qu phc tp):", mse_complex)
```

```python
from sklearn.linear_model import Ridge, Lasso

# S dng Ridge Regression  kim sot overfitting
ridge_model = Ridge(alpha=0.01)  # Thay i gi tr alpha  iu chnh mc  regularization
ridge_model.fit(X_poly, y)
y_pred_ridge = ridge_model.predict(X_test_poly)

# S dng Lasso Regression  kim sot overfitting
lasso_model = Lasso(alpha=0.01)  # Thay i gi tr alpha  iu chnh mc  regularization
lasso_model.fit(X_poly, y)
y_pred_lasso = lasso_model.predict(X_test_poly)

# V biu  d liu v d on ca c hai m hnh
plt.scatter(X, y, s=20, label='D liu thc t')
plt.plot(X_test, y_pred_ridge, label='M hnh Ridge Regression', color='g')
plt.plot(X_test, y_pred_lasso, label='M hnh Lasso Regression', color='r')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('S dng Ridge v Lasso  kim sot overfitting')
plt.show()

# o  tt ca m hnh bng Mean Squared Error
mse_ridge = mean_squared_error(y, ridge_model.predict(X_poly))
mse_lasso = mean_squared_error(y, lasso_model.predict(X_poly))

print("MSE (Ridge):", mse_ridge)
print("MSE (Lasso):", mse_lasso)
```
